```python
#Aiming to draw a boundary around the leaf, so that that the points outside the boundarycan be made black .

img = Image.open(r"C:\Users\Revanna\Desktop\plant disease test_dataset\testSet\potato___healthy.jpg")
img1 = Image.open(r"C:\Users\Revanna\Desktop\plant disease test_dataset\testSet\Apple___Apple_scab.jpg")
len_bound_points = 0
len_bound_points1 = 0
t = 5
for x in range(t,img.width-t):
        for y in range(t,img.height-t):
            r,g,b=img.getpixel((x,y))
            rl,gl,bl=img.getpixel((x-t,y))
            rb,gb,bb=img.getpixel((x,y+t))
            rr,gr,br=img.getpixel((x+t,y))
            rt,gt,bt=img.getpixel((x,y-t))
            if(g>r and g>b):
                if((gl<rl and gl<bl) or (gb<rb and gb<bb) or (gb<rb and gb<bb) or (gt<rt and gt<bt)):
                    img.putpixel((x,y), (250,250,250))
                    len_bound_points += 1
                    
            r,g,b=img1.getpixel((x,y))
            rl,gl,bl=img1.getpixel((x-t,y))
            rb,gb,bb=img1.getpixel((x,y+t))
            rr,gr,br=img1.getpixel((x+t,y))
            rt,gt,bt=img1.getpixel((x,y-t))
            if(g>r and g>b):
                if((gl<rl and gl<bl) or (gb<rb and gb<bb) or (gb<rb and gb<bb) or (gt<rt and gt<bt)):
                    img1.putpixel((x,y), (250,250,250))
                    len_bound_points1 += 1
display(img,len_bound_points)
print("-----------------------------------------------------")
display(img1,len_bound_points1)
```
